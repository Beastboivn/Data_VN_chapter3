## GDP Deflator Dataset Cleaning

```python
# First, we import DGDP dataset which was downloaded from the link: https://databank.worldbank.org/reports.aspx?source=2&series=NY.GDP.DEFL.ZS&country=#
dgdp = pd.read_csv("/content/dgdp.csv")
dgdp.head()
```

```python
# Next, we rename the year columns to make them look cleaner
new_cols = {}
for col in dgdp.columns:
    if '[' in col and 'YR' in col:
        # extract year before space
        year = col.split()[0]
        new_cols[col] = year
dgdp = dgdp.rename(columns=new_cols)
dgdp.head()
```

```python
# Next, we rebase GDP deflator to 2015
dgdp2 = dgdp.iloc[:-5,:]
multipliers = 100 / dgdp2['2015']
for i in dgdp2.iloc[:,4:].columns:
  dgdp2.loc[:,i] = dgdp2.loc[:,i]*multipliers

dgdp2.head()
```

```python
# Finally, we transform the wide dataframe to the long format for easier future use:
dgdp_clean = dgdp2.iloc[:,2:].melt(id_vars={'Country Name', 'Country Code'}, var_name= 'year', value_name= 'dgdp')
dgdp_clean = dgdp_clean.rename(columns = {'Country Name': 'country', 'Country Code': 'country_code'})
dgdp_clean['year'] = dgdp_clean['year'].astype(int)
dgdp_clean
```

```python
# However, the dataset above is not converted to 2015 constant USD yet.
# To do that, we first merge the appended dataframe with the clean PPP and DGDP datasets:

appended_df2 = appended_df.merge(ppp_clean, on = ['country_code', 'country']).merge(dgdp_clean,
                                                                                    on = ['country_code', 'country', 'year'])

# The code below convert OBS_VALUE to 2015 constant USD OBS_VALUE
appended_df2['OBS_VALUE'] = appended_df2['OBS_VALUE']* 100 / (appended_df2['ppp']*appended_df2['dgdp'])
appended_df2
```

## CO2 Dataset Cleaning

```python
# First , we import the CO2 dataset, which was downloaded from the link: https://data-explorer.oecd.org/vis?tm=DF_ICIO_GHG_EXPD_2023&snb=5&df[ds]=dsDisseminateFinalDMZ&df[id]=DSD_ICIO_GHG_SCOPE%2540DF_ICIO_GHG_SCOPE_2023&df[ag]=OECD.STI.PIE&df[vs]=1.0&dq=A.SCOPE_GHG.AUT%2BBEL%2BCAN%2BCZE%2BEST%2BFIN%2BFRA%2BGRC%2BIRL%2BITA%2BJPN%2BLVA%2BLTU%2BLUX%2BNLD%2BNOR%2BSVK%2BSVN%2BSWE%2BGBR%2BUSA.A01_02%2BA03%2BG%2BH49%2BH50%2BH51%2BH52%2BO%2BN%2BM%2BL%2BK%2BJ62_63%2BJ61%2BJ58T60%2BI%2BH53%2BP%2BQ%2BR%2BS%2BT%2BC10T12%2BC13T15%2BC16%2BC17_18%2BC19%2BC20%2BC21%2BC22%2BC23%2BC24%2BC25%2BC26%2BC27%2BC28%2BC29%2BC30%2BC31T33._T.&pd=2000%2C2020&to[TIME_PERIOD]=false&vw=ov&lb=bt
co2 = pd.read_csv("/content/co2.csv")

# Next, we select necessary columns:
co2 = co2[['EMISSIONS_ORIGIN_AREA', 'Emissions origin area', 'ACTIVITY', 'Economic activity', 'TIME_PERIOD', 'OBS_VALUE']]
co2
```

```python
# After that, we classify the activities into 3 main industries:
co2['Economic activity'] = co2['ACTIVITY'].str[0].map({
    'A': 'Agriculture, forestry and fishing',
    'C': 'Manufacturing'
}).fillna('Services')
co2
```

```python
# Next, we calculate the aggregated total of CO2 emissions:
co2 = co2.groupby(['EMISSIONS_ORIGIN_AREA',	'Emissions origin area', 'Economic activity', 'TIME_PERIOD'])['OBS_VALUE'].sum().reset_index()
co2['Measure'] = 'co2'
co2_clean = co2[['EMISSIONS_ORIGIN_AREA',	'Emissions origin area', 'Economic activity', 'Measure', 'TIME_PERIOD', 'OBS_VALUE']]
co2_clean = co2_clean.rename(columns = {'EMISSIONS_ORIGIN_AREA' : 'country_code', 'Emissions origin area': 'country', 'TIME_PERIOD': 'year'})

# Finally, we multiply the OBS_VALUE by 1000 to convert it to thousands tonnes of CO2 equilvalent.
co2_clean['OBS_VALUE'] = co2_clean['OBS_VALUE'] * 1000
co2_clean.to_csv('co2.csv', index = False)
```

## Making Data Description Table

### Making 'Statistical Description' Table

```python
# Let's take another quick look at our gtfp variables dataframe:
gtfp_vars_clean.head()
```

```python
# In this research, we only use the variable K for GTFP calculation.
# To compare the effect of ICT capital and non-ICT capital, we need to create the 'non-ict K' variable:
gtfp_vars_clean['non_ict_k'] = gtfp_vars_clean['k'] - gtfp_vars_clean['ict']
gtfp_vars_clean.head()
```

```python
# We want our 'Statistical Description' table to be formatted as follow:
# 7 columns: Sector, Variables, Mean, Median, Std, Max, Min
# To do that, the idea is to first unpivot the 'gtfp_vars_clean' table and use the 'groupby()' function.

# Step 1: unpivot the table:
gtfp_unpivot = gtfp_vars_clean.melt(id_vars = {'ref_area', 'referencearea', 'economicactivity', 'year'}, var_name = 'variables', value_name = 'value')
# We remove the 'k' variable because we don't need it anymore.
gtfp_unpivot = gtfp_unpivot[gtfp_unpivot['variables'] != 'k']

# Step 2: aggregate the table to get the expected results.
gtfp_unpivot = gtfp_unpivot.groupby(['economicactivity', 'variables'])['value'].agg(['mean', 'median', 'std', 'max', 'min']).reset_index().round(2)
gtfp_unpivot
```
